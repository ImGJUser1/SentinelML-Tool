```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sentinelml.core import Sentinel

X, y = load_iris(return_X_y=True)

model = RandomForestClassifier().fit(X, y)

sentinel = Sentinel()
sentinel.fit(X)
```

```python
scores = []

for x in X:
    scores.append(sentinel.assess(x)["trust"])
```

```python
from sentinelml.viz import plot_trust

plot_trust(scores)
```

```python
import numpy as np

drift_data = X + np.random.normal(0,2,X.shape)

for x in drift_data[:20]:
    print(sentinel.assess(x))
```
